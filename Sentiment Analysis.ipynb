```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('Train.csv')
```

```python
df
```

```python
c = []
for sent in df['text']:
    c.append(len(sent.split()))
df['count'] = c    
```

```python
l = []
for sent in df['text']:
    if len(sent.split())<61:
        l.append(sent)
    else:
        start = sent.split()[:30]
        start = ' '.join(start)
        end = sent.split()[len(sent.split())-31:]
        end = ' '.join(end)
        l.append(start+end)
```

```python
df['truncated_text'] = l
```

```python
df
```

```python
c = []
for sent in df['truncated_text']:
    c.append(len(sent.split()))
df['new_count'] = c    
```

```python
df
```

```python
df.drop(['text','count','new_count'],axis=1,inplace=True)
```

```python
df
```

```python
import re
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
sentences = []
for sent in df['truncated_text']:
    sent = re.sub('[^a-zA-Z]', ' ',sent)
    sent = sent.lower()
    sent = sent.split()
    sent = [ lemmatizer.lemmatize(word) for word in sent if word not in set(stopwords.words('english')) ]
    sent = ' '.join(sent)
    sentences.append(sent)
df['new_text'] = sentences 
```

```python
df
```

```python
c = []
for sent in df['new_text']:
    c.append(len(sent.split()))
df['new_count'] = c    
```

```python
df
```

```python
df.drop(['truncated_text','new_count'],axis=1,inplace=True)
```
